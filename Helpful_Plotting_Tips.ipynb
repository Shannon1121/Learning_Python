```python
# Labelling plots with a legend vs in-line

# Link to a github repository where someone made a function allowing for plots to be labelled 'in-line', rather than a separate box to the side (see images in repository)
    # https://github.com/cphyc/matplotlib-label-lines
# REQUIRED: Only install once
# pip install matplotlib-label-lines 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import loglaplace,chi2
from labellines import labelLine, labelLines

# Some math heavy plotting as an example. Same on the github URL linked
X = np.linspace(0, 1, 500)
A = [1, 2, 5, 10, 20]
funcs = [np.arctan, np.sin, loglaplace(4).pdf, chi2(5).pdf]

for a in A:
    plt.plot(X, np.arctan(a*X), label=str(a))
    
# z-order will change the back/front occupancy of the text. Play around with it. 
labelLines(zorder = 5)
# labelLines(zorder = 0.5)
```
